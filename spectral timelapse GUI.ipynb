```python
import numpy as np
import time, os

timelapse_dir = "/media/martin/data/SpecimIQ_images/banana_experiment"

dirs = os.listdir(timelapse_dir)
dirs.sort() #not sorted from os.listdir

def get_specim(rawfile):
    f = open(rawfile, "r")
    im = np.fromfile(f, dtype=np.float32)
    
    im = np.reshape(im,[512, 204, 512])
    im = np.transpose(im,(2,0,1))[:,::-1,:] #reshape to 512x512x204 with correct orientation

    #normalize to 0 to 255 ints
    return np.round(im/np.max(im)*255).astype('uint8')

if True:
    #preprocess data to memorymaps
    for curdir in dirs:
        print(curdir)
        fp = np.memmap(timelapse_dir+'/'+curdir+'/mmap', dtype='uint8', mode='w+', shape=(512,512,204))
        im = get_specim(rawfile = timelapse_dir+"/"+curdir+'/results/REFLECTANCE_'+curdir+'.dat')
        fp[:] = im[:]
        del fp #Deletion flushes memory changes to disk before removing the object
```
